```python
import pandas as pd
```

```python
import re
```

```python
f=open('Group Chat.txt', 'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
pattern= '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[A-Z]{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
len(messages)
```

```python
dates = re.findall(pattern, data)
```

```python
len(dates)
```

```python
def remove(elements,item):
    return [element.replace(item, "") for element in elements]

dates=remove(dates,'\u202f')
```

```python
len(dates)
```

```python
def strip(elements):
    return [element.rstrip(" - ") for element in elements]

dates=strip(dates)
```

```python
dates[5]
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format="%m/%d/%y, %I:%M%p")
# df['message_date'] = df['message_date'].dt.strftime("%m/%d/%y, %H:%M ")
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df
```

```python
    users = []
    messages = []
    for message in df['user_message']:
        entry = re.split('([\w\W]+?):\s', message)
        if entry[1:]:  # user name
            users.append(entry[1])
            messages.append(" ".join(entry[2:]))
        else:
            users.append('group_notification')
            messages.append(entry[0])

    df['user'] = users
    df['message'] = messages
    df.drop(columns=['user_message'], inplace=True)
```
